# **Laboratorio 3: Una introducción a Matplotlib**


## **1. Gráficos simples**

Para comenzar la introducción a la librería *Matplotlib*, deben graficar curvas simples y conjuntos de datos en un solo gráfico con los respectivos nombres de los ejes, título, leyendas y escalas. En general, para este punto, deben importar *pyplot*, de la siguiente forma:



```
import matplotlib.pyplot as plt
```

**1.1. Para ejemplificar la forma de realizar gráficos simples, se grafican las funciones $sin(t)$ y $cos(t)$, donde $t$ es el tiempo. Para esto, se crea primero un arreglo con que contenga los datos de tiempo**

````python
import numpy as np

t = np.linspace(0,10,100)
````

````python
import matplotlib.pyplot as plt
````

````python
##METODO 1
plt.figure()
plt.plot(t,np.sin(t),"r",label="Seno")
plt.plot(t,np.cos(t),label="Coseno")
plt.grid()
plt.xlabel("t(seg)")
plt.ylabel("f(t)")
plt.title("Metodo 1")
plt.legend()
plt.show()
````

````python

##METODO 2

fig, ax = plt.subplots()
ax.plot(t,np.sin(t),label="Seno")
ax.set_title("Seno")

ax2 = ax.twinx()
ax2.plot(t,np.cos(t),label="Coseno")
"""ax.set_title("Seno")
ax.set_xlabel("t(seg)")"""
fig.suptitle("Funciones trigonometricas")

"""ax.grid()
ax.set_xlabel("t(seg)")
ax.set_ylabel("f(t)")
ax.set_title("Metodo 1")
ax.legend()
ax.grid()"""
fig.set_tight_layout(True)
plt.show()
````

Luego, se usa la librería *Numpy* para obtener los datos en la imágen de las funciones mencionadas

````python
s = np.sin(t)
c = np.cos(t)
````

Ahora, se debe conocer la función *subplots()*, la cual permite retornar los atributos *fig* y *ax* con diversos parámetros para poder configurar el gráfico. Algunos parámetros de esta función son:

- *nrows*, *ncols*: Número de filas y columnas en una cuadrícula de subplots (Valor 1 por defecto)
- *sharex, sharey*: Permite compartir eje *x* y eje *y* (Valor por defecto 'False')
- *figsisze*: Permite establecer un tamaño del gráfico. *figsize = (Height, width)*
- *layout* = Ajusta la cuadrícula para evitar que se sobrelapen los sublots. *layout = 'constrained'*

La forma de usar esta función es:

```
fig, ax = plt.subplots(nrows, ncols, sharex = False/True, sharey = False/True, figsize = (height, width))
```


Debido a que solo hay un subplot, el atributo *ax* solo tiene un elemento que corresponde a dicho subplot. Luego, para graficar lo anterior, a este atributo, es posible aplicarle algunas funciones para graficar. Algunas de ellas son:


- *ax.plot(x,y,label = 'leyenda para la curva', linewidth = Grosor de curva)* : Grafica el conjunto de datos *y* contra el conjunto *x*
- *ax.set(xlabel = 'Título del eje x', ylabel = 'Título del eje y', title = 'Título del subplot')*: Asigna títulos a los ejes y al subplot
- *ax.legend(loc = 'Ubicación de la leyenda')*: Muestra la leyenda de las curvas graficadas, la ubicaciones pueden ser: *'right', 'left', 'center', 'upper right', 'upper left', 'lower right', 'lower left', 'center right', 'center left', 'upper center, 'lower center'*
- *ax.grid()*: Muestra una cuadrícula.
- *ax.set_xlim([mínimo, máximo])*: Establece límites para el eje $x$.
- *ax.set_ylim([mínimo, máximo])*: Establece límites para el eje $y$.

Cabe resaltar que hay otras formas de configurar el gráfico para llegar al mismo resultado, depende del estilo de cada programador.

Luego, para darle un título general al gráfico, se hace de la forma:

```
fig.suptitle('Título general del gráfico')
```
Finalmente, para mostrar el gráfico, se escribe:

```
plt.show()
```
A continuación, se muestra el ejemplo anterior:


````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(t,s,label='Función seno')
ax.plot(t,c,label='Función coseno')
ax.set(xlabel = 'Tiempo [s]', ylabel='Funciones', title = 'Seno vs Coseno')
ax.legend(loc='lower right')
fig.suptitle('Gráfico simple')
plt.show()
````

Si se quisera presentar las curvas en dos subplots diferentes, hay que tener en cuenta que al cambiar el número de filas y/o columnas, el atributo *ax* cambia su tamaño para ser un arreglo. Por ejemplo, si se quisiera presentar un gráfico de dos filas y dos columnas, la forma sería:

```
fig, ax = plt.subplots(2, 2)
```

Donde *ax* ahora es un arreglo de tamaño $2 \times 2$, luego, si se quiere llamar algún subplot en específico, por ejemplo, el que se encuentra en la posición $(2,1)$, se hace de la forma:

```
ax[1,0]
```

A continuación, se muestra la forma en la que se presentarían las curvas anteriroes en dos subplots:

````python
fig, ax = plt.subplots(1 , 2)

ax[0].plot(t, s)
ax[0].set_title('Función Seno')
ax[0].set(xlabel = 'Tiempo [s]', ylabel='Funciones')


ax[1].plot(t, s)
ax[1].set_title('Función Coseno')
ax[1].set(xlabel = 'Tiempo [s]', ylabel='Funciones')
ax[1].label_outer()

fig.suptitle('Gráfico simple')
plt.show()
````

Es necesario aclarar que hay diferentes formas de graficar un conjunto de datos. Para esto, se remplaza la función *plot* por otras como:

- *scatter*: Grafica los conjuntos de datos como puntos.
- *bar*: Grafica los conjuntos como barras.
- *errorbar*: Incluye barras de error para datos estadísticos.
- *fill_between*: Sobrea áreas de confianza.
- *hist*: Muestra un histograma de los datos.

La documentación de estas funciones se encuentra en el siguiente link: [Documentación](https://matplotlib.org/stable/gallery/index.html)
