## **1. Grficos simples**

Para comenzar la introduccin a la librera *Matplotlib*, deben graficar curvas simples y conjuntos de datos en un solo grfico con los respectivos nombres de los ejes, ttulo, leyendas y escalas. En general, para este punto, deben importar *pyplot*, de la siguiente forma:



```
import matplotlib.pyplot as plt
```

**1.1. Para ejemplificar la forma de realizar grficos simples, se grafican las funciones $sin(t)$ y $cos(t)$, donde $t$ es el tiempo. Para esto, se crea primero un arreglo con que contenga los datos de tiempo**

````python
import numpy as np

t = np.linspace(0,10,100)
````

````python
import matplotlib.pyplot as plt
````

````python
##METODO 1
plt.figure()
plt.plot(t,np.sin(t),"r",label="Seno")
plt.plot(t,np.cos(t),label="Coseno")
plt.grid()
plt.xlabel("t(seg)")
plt.ylabel("f(t)")
plt.title("Metodo 1")
plt.legend()
plt.show()
````

````python

##METODO 2

fig, ax = plt.subplots()
ax.plot(t,np.sin(t),label="Seno")
ax.set_title("Seno")

ax2 = ax.twinx()
ax2.plot(t,np.cos(t),label="Coseno")
"""ax.set_title("Seno")
ax.set_xlabel("t(seg)")"""
fig.suptitle("Funciones trigonometricas")

"""ax.grid()
ax.set_xlabel("t(seg)")
ax.set_ylabel("f(t)")
ax.set_title("Metodo 1")
ax.legend()
ax.grid()"""
fig.set_tight_layout(True)
plt.show()
````

Luego, se usa la librera *Numpy* para obtener los datos en la imgen de las funciones mencionadas

````python
s = np.sin(t)
c = np.cos(t)
````

Ahora, se debe conocer la funcin *subplots()*, la cual permite retornar los atributos *fig* y *ax* con diversos parmetros para poder configurar el grfico. Algunos parmetros de esta funcin son:

- *nrows*, *ncols*: Nmero de filas y columnas en una cuadrcula de subplots (Valor 1 por defecto)
- *sharex, sharey*: Permite compartir eje *x* y eje *y* (Valor por defecto 'False')
- *figsisze*: Permite establecer un tamao del grfico. *figsize = (Height, width)*
- *layout* = Ajusta la cuadrcula para evitar que se sobrelapen los sublots. *layout = 'constrained'*

La forma de usar esta funcin es:

```
fig, ax = plt.subplots(nrows, ncols, sharex = False/True, sharey = False/True, figsize = (height, width))
```


Debido a que solo hay un subplot, el atributo *ax* solo tiene un elemento que corresponde a dicho subplot. Luego, para graficar lo anterior, a este atributo, es posible aplicarle algunas funciones para graficar. Algunas de ellas son:


- *ax.plot(x,y,label = 'leyenda para la curva', linewidth = Grosor de curva)* : Grafica el conjunto de datos *y* contra el conjunto *x*
- *ax.set(xlabel = 'Ttulo del eje x', ylabel = 'Ttulo del eje y', title = 'Ttulo del subplot')*: Asigna ttulos a los ejes y al subplot
- *ax.legend(loc = 'Ubicacin de la leyenda')*: Muestra la leyenda de las curvas graficadas, la ubicaciones pueden ser: *'right', 'left', 'center', 'upper right', 'upper left', 'lower right', 'lower left', 'center right', 'center left', 'upper center, 'lower center'*
- *ax.grid()*: Muestra una cuadrcula.
- *ax.set_xlim([mnimo, mximo])*: Establece lmites para el eje $x$.
- *ax.set_ylim([mnimo, mximo])*: Establece lmites para el eje $y$.

Cabe resaltar que hay otras formas de configurar el grfico para llegar al mismo resultado, depende del estilo de cada programador.

Luego, para darle un ttulo general al grfico, se hace de la forma:

```
fig.suptitle('Ttulo general del grfico')
```
Finalmente, para mostrar el grfico, se escribe:

```
plt.show()
```
A continuacin, se muestra el ejemplo anterior:


````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(t,s,label='Funcin seno')
ax.plot(t,c,label='Funcin coseno')
ax.set(xlabel = 'Tiempo [s]', ylabel='Funciones', title = 'Seno vs Coseno')
ax.legend(loc='lower right')
fig.suptitle('Grfico simple')
plt.show()
````

Si se quisera presentar las curvas en dos subplots diferentes, hay que tener en cuenta que al cambiar el nmero de filas y/o columnas, el atributo *ax* cambia su tamao para ser un arreglo. Por ejemplo, si se quisiera presentar un grfico de dos filas y dos columnas, la forma sera:

```
fig, ax = plt.subplots(2, 2)
```

Donde *ax* ahora es un arreglo de tamao $2 \times 2$, luego, si se quiere llamar algn subplot en especfico, por ejemplo, el que se encuentra en la posicin $(2,1)$, se hace de la forma:

```
ax[1,0]
```

A continuacin, se muestra la forma en la que se presentaran las curvas anteriroes en dos subplots:

````python
fig, ax = plt.subplots(1 , 2)

ax[0].plot(t, s)
ax[0].set_title('Funcin Seno')
ax[0].set(xlabel = 'Tiempo [s]', ylabel='Funciones')


ax[1].plot(t, s)
ax[1].set_title('Funcin Coseno')
ax[1].set(xlabel = 'Tiempo [s]', ylabel='Funciones')
ax[1].label_outer()

fig.suptitle('Grfico simple')
plt.show()
````

Es necesario aclarar que hay diferentes formas de graficar un conjunto de datos. Para esto, se remplaza la funcin *plot* por otras como:

- *scatter*: Grafica los conjuntos de datos como puntos.
- *bar*: Grafica los conjuntos como barras.
- *errorbar*: Incluye barras de error para datos estadsticos.
- *fill_between*: Sobrea reas de confianza.
- *hist*: Muestra un histograma de los datos.

La documentacin de estas funciones se encuentra en el siguiente link: [Documentacin](https://matplotlib.org/stable/gallery/index.html)
